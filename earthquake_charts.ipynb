```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
#where the data came from ?

# this data set is taken from US Geological Survey
# the dataset contains details of all
# the earthquakes that have happend in the last 30 days
# it gets updated every 15 minutes
```

```python
# I want to answer these three questions in my Data Exploratory analysis:

# 1) What is the distribution of earthquakes in the last 30 days?
# 2) Which location has the highest number of earthquakes?
# 3) What is the average magnitude of earthquakes per location?
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

data_set = pd.read_csv("datasets/earthquakes_coordinates.csv")

countries = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
countries.plot(color="black", ax=ax)

data_set.plot(x="longitude", y="latitude", kind="scatter",colormap="YlOrRd", ax=ax)

plt.xlabel('Latitude', fontsize=14, fontweight='bold')
plt.ylabel('Longitude', fontsize=14, fontweight='bold')
ax.set_title("Earthquake Distribution in the last 30 days", fontsize=16, fontweight='bold')

ax.grid(visible=False)
fig.tight_layout()
```

```python
# 1) What is the distribution of earthquakes in the last 30 days?

# In the graph above, I have plotted all the coordinates of longitude and latitude,
# showing the distribution of earthquakes in the last 30 days.
# Each blue point represents the location of an earthquake coordinate.
# Therefore, areas with more blue points indicate a higher frequency of earthquakes in those regions.
# most earthquakes occur at the plate boundaries
# mostly concentrated in north amercia and asia pacific
```

```python
data_set = pd.read_csv("datasets/earthquake_count_and_location_data.csv")
fig, ax = plt.subplots(figsize=(15, 10))
ax.barh(data_set["location"], data_set["count"])
plt.gca().invert_yaxis()

plt.xlabel('Number of Earthquakes', fontsize=14, fontweight='bold')
plt.ylabel('Locations', fontsize=14, fontweight='bold')

ax.set_title("Top 30 High Impact Locations", fontsize=16, fontweight='bold')

fig.tight_layout()
```

```python
# 2) Which location has the highest number of earthquakes?

# In this graph, the top 30 locations with the highest number of earthquakes are displayed,
# with the location having the highest number of earthquakes positioned at the top.
```

```python
data_set = pd.read_csv("datasets/earthquake_median_per_location.csv")

fig, ax = plt.subplots(figsize=(15, 10))
plt.xticks(rotation=90)

plt.xlabel('Locations', fontsize=14, fontweight='bold')
plt.ylabel('Magnitude', fontsize=14, fontweight='bold')
ax.set_title("Magnitudes over the Last 30 Days", fontsize=16, fontweight='bold')

ax.plot(data_set["locations"], data_set["median_magnitude"])
```
